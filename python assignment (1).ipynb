```python
# Q.14 Write a programme to reverse string using a while loop.
string= input("please enter string  ")
lenn = len(string)
reverse = ""
while lenn > 0:
  reverse = reverse + string[lenn-1]
  lenn = lenn - 1
print(reverse)
```

```python
# Q.15 Write the programme to calculate the factorial of number provided by user using while loop.
num = int(input("please enter a number  "))
fact = 1
while num > 0:
  fact = fact * num
  num = num - 1
print(fact)
```
